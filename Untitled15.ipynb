```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from mpl_toolkits.mplot3d import Axes3D

data = {
    'Feature1': [2, 4, 6, 8, 10],
    'Feature2': [1, 3, 5, 7, 9],
    'Feature3': [5, 10, 15, 20, 25]
}

df = pd.DataFrame(data)

scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

fig = plt.figure(figsize=(12, 5))
ax1 = fig.add_subplot(121, projection='3d')
ax1.set_xlabel('Feature1')
ax1.set_ylabel('Feature2')
ax1.set_zlabel('Feature3')
ax1.set_title('Before PCA (3d)')
ax1.legend()

ax1.scatter(df['Feature1'], df['Feature2'], df['Feature3'], color='blue', label='Original Data')

pca = PCA(n_components=2)
principalComponents = pca.fit_transform(scaled_data)

pca_df = pd.DataFrame(data = principalComponents, columns = ['PC1', 'PC2'])

ax2 = fig.add_subplot(122)
ax2.scatter(pca_df['PC1'], pca_df['PC2'], color='red', label='PCA Transformed data')
ax2.set_xlabel('Principal Amount 1')
ax2.set_ylabel('Principal Amount 2')
ax2.set_title('After PCA (2d)')
ax2.legend()

plt.tight_layout()
plt.show()
```
